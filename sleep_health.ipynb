# Sleep and Health Classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
print(df.shape)
df.head()
```

# Exploratory Analysis

Separate Blood Pressure Values into high and low values

```python
df["High_pressure"] = 0
df["Low_pressure"] = 0

for i, val in enumerate(df["Blood Pressure"]):
    j = 0

    while val[j] != "/":
        j += 1
        continue

    df.loc[i, "High_pressure"] = int(val[:j])
    df.loc[i, "Low_pressure"] = int(val[(j+1):])

df = df.drop("Blood Pressure", axis=1)
df.head()
```

Numerical Variable Distributions

```python
fig, axs = plt.subplots(3, 3, figsize=(8, 8))
plt.subplots_adjust(wspace=0.8, hspace=0.8)

numeric = ["Age", "Sleep Duration", "Quality of Sleep", "Physical Activity Level", "Stress Level", "Heart Rate", "Daily Steps", "High_pressure", "Low_pressure"]

for i, cat in enumerate(numeric):
    sb.histplot(df[cat], bins=10, color='skyblue', ax=axs[int(i/3), i%3]).set_ylabel("")

plt.show()
```

Categorical Variable Distributions

```python
categorical = ['Gender', 'Occupation', 'BMI Category', 'Sleep Disorder']

fig, axs = plt.subplots(2, 2, figsize=(10, 6))
plt.subplots_adjust(wspace=0.25, hspace=1)

for i, cat in enumerate(categorical):
    sb.countplot(x=cat, data=df, color='skyblue', ax=axs[int(i/2), i%2]).set_ylabel("")
    plt.xticks(rotation=45)

for ax in axs.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(45)
        label.set_ha('right')

plt.tight_layout()
plt.show()
```

```python
df['BMI Category'].value_counts()
```

Convert Normal Weight Category to Normal, and combine Overweight and Obese categories, since the model may not be very reliable if only 10 samples are available.

```python
df.loc[df['BMI Category'] == 'Normal Weight', 'BMI Category'] = 'Normal'
df.loc[df['BMI Category'] == 'Obese', 'BMI Category'] = 'Overweight'
df['BMI Category'].value_counts()
```

Look at Gender and BMI frequency chart:

```python
pd.crosstab(df['Gender'], df['BMI Category'], margins=True)
```

Sleep disorder and BMI:

```python
pd.crosstab(df['Sleep Disorder'], df['BMI Category'], margins=True)
```

Do the distributions in sleep time differ between these three sleep disorder categories?

```python
fig, axs = plt.subplots(1, 3, figsize=(6, 3))
index = 0
for disorder in pd.unique(df['Sleep Disorder']):
    sub = df[df['Sleep Disorder'] == disorder]
    axs[index].hist(sub['Sleep Duration'])
    axs[index].set_title(disorder)
    index += 1
plt.show()
```

As expected, people with insomnia tend to sleep less at night. Sleep duration for people with sleep apnea is skewed at both extremes.

Correlation Heatmap

```python
sb.heatmap(df.drop('Person ID', axis=1).corr(numeric_only=True), annot=True)
plt.show()
```

Strong correlations (> 0.6 or <-0.6) between Sleep duration and sleep quality, daily steps and physical activity, heart rate and stress level, sleep duration and stress level, and heart rate and sleep quality.

# BMI Classification

Question: Can we predict an individual's BMI category based on these seven measures?

Split into training and testing sets.

```python
from sklearn.model_selection import train_test_split

X = df[['Age', 'Sleep Duration', 'Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps']]
y = df['BMI Category']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
import time

times = []

# Create a logistic regression model
model = LogisticRegression(max_iter=200)

# Fit the model to the training data
start = time.time()
model.fit(X_train, y_train)
end = time.time()
print('Training Time:', end-start)
times.append(end-start)

# Predict the labels of the test data
y_pred = model.predict(X_test)
y_train_predict = model.predict(X_train)
print("Training Set Accuracy: %d / %d = %f"
      % ((y_train == y_train_predict).sum(), X_train.shape[0], model.score(X_train, y_train)))
# Test set accuracy
print("Testing Set Accuracy: %d / %d = %f"
      % ((y_test == y_pred).sum(), X_test.shape[0], model.score(X_test, y_test)))
```

```python
from sklearn.model_selection import KFold, cross_val_score

# Create a KFold object with 5 folds
kf = KFold(n_splits=5)
accuracy = []

scores_lr = cross_val_score(model, X, y, cv = kf)

print("Cross Validation Scores: ", scores_lr)
print("Average CV Score: ", scores_lr.mean())
accuracy.append(scores_lr.mean())
print("Number of CV Scores used in Average: ", len(scores_lr))
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

# Gaussian Naive Bayes
gnb = GaussianNB()
start = time.time()
model = gnb.fit(X_train, y_train)
end = time.time()
print('Training Time:', end-start)
times.append(end-start)

# Training set accuracy
y_train_predict = model.predict(X_train)
print("Training Set Accuracy: %d / %d = %f"
      % ((y_train == y_train_predict).sum(), X_train.shape[0], model.score(X_train, y_train)))
# Test set accuracy
y_pred = model.predict(X_test)
print("Training Set Accuracy: %d / %d = %f"
      % ((y_test == y_pred).sum(), X_test.shape[0], model.score(X_test, y_test)))
```

Kfold cross-validation

```python
scores = cross_val_score(gnb, X, y, cv = kf)

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
accuracy.append(scores.mean())
print("Number of CV Scores used in Average: ", len(scores))
```

## SVM

```python
from sklearn.svm import SVC

# Create the SVM classifier
clf = SVC()

# Train the classifier
start = time.time()
model = clf.fit(X_train, y_train)
end = time.time()
print('Training Time:', end-start)
times.append(end-start)

# Training set accuracy
y_train_predict = model.predict(X_train)
print("Training Set Accuracy: %d / %d = %f"
      % ((y_train == y_train_predict).sum(), X_train.shape[0], model.score(X_train, y_train)))
# Test set accuracy
y_pred = model.predict(X_test)
print("Training Set Accuracy: %d / %d = %f"
      % ((y_test == y_pred).sum(), X_test.shape[0], model.score(X_test, y_test)))
```

```python
scores_svm = cross_val_score(clf, X, y, cv = kf)

print("Cross Validation Scores: ", scores_svm)
print("Average CV Score: ", scores_svm.mean())
accuracy.append(scores_svm.mean())
print("Number of CV Scores used in Average: ", len(scores_svm))
```

## Conclusion

```python
print(accuracy)
print(times)
```

```python
fig, ax1 = plt.subplots()
fig.suptitle('Classification Model Comparisons')
ax1.bar(range(len(accuracy)), accuracy, align='edge', width=-0.3, color='green')
ax1.set_ylabel('Testing Set Accuracy')
ax1.set_xticks(range(3), ['Logistic Regression', 'Naive Bayes', 'SVM'])
ax1.set_xlabel('Classification Model')
ax2 = ax1.twinx()
ax2.bar(range(len(times)), times, align='edge', width=0.3, color='red')
ax2.set_ylabel('Training Time (sec)')
ax2.grid(visible=False)
fig.legend(['Model Accuracy', 'Training Time'])

plt.show()
```

Logistic Regression appears to have the best accuracy. However, it also takes the longest to train. In this case, the training times for all 3 models are all very small, so this difference is negligible, and accuracy should be prioritized. For larger datasets however, the difference in training time may become significant.
